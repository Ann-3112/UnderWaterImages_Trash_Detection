```python
from ultralytics import YOLO

# Load model
model = YOLO("yolov8n.pt")

# Train
# We can use higher settings now because Colab is powerful!
model.train(
    data="/content/data.yaml",  # Path to the config we unzipped
    epochs=25,                  # 25 epochs is a great balance
    imgsz=640,                  # Full HD training
    batch=16,                   # Higher batch size
    name="my_trash_model"
)
```

```python
import shutil
import os

# Source (Where Colab saved it)
source = "/content/runs/detect/my_trash_model/weights/best.pt"

# Destination (Your Google Drive)
destination = "/content/drive/My Drive/TrashAI/best.pt"

if os.path.exists(source):
    shutil.copy(source, destination)
    print(f"✅ SUCCESS! Model saved to Google Drive at: {destination}")
    print("Go download it now!")
else:
    print("❌ Error: Could not find the trained model.")
```
